## FRAUD CLEANING

```python
df_fraud = df
```

```python
df_fraud['Time'] = df_fraud['Time'] / 3600
df_fraud.rename(columns={'Time': 'Time_hours'}, inplace=True)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
df_fraud['Amount_scaled'] = scaler.fit_transform(df_fraud[['Amount']])
df_fraud = df_fraud.drop('Amount', axis=1)
```

```python
print(f"Shape: {df_fraud.shape}")
print(f"Missing values: {df_fraud.isnull().sum().sum()}")
print(f"Fraud rate: {df_fraud['Class'].mean()*100:.3f}%")
print(f"\nTime range (hours): {df_fraud['Time_hours'].min():.1f} - {df_fraud['Time_hours'].max():.1f}")
print(f"Amount_scaled mean: {df_fraud['Amount_scaled'].mean():.4f}")
```
