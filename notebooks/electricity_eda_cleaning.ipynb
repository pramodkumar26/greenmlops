```python
# COMBINE ALL 12 MONTHLY FILES
import pandas as pd
import glob
import os
import openpyxl

files = glob.glob('C:\\IP\\greenmlops\\data\\raw\\electricity\\*.xlsx')

# Load and combine all files
dfs = []
for f in files:
    temp = pd.read_excel(f)
    dfs.append(temp)
    filename = os.path.basename(f)  # ‚Üê fix: extract filename separately
    print(f"Loaded {filename}: {temp.shape}")

# Combine into one dataframe
df_elec = pd.concat(dfs, ignore_index=True)

# First look
print(f"\n--- COMBINED SHAPE: {df_elec.shape} ---")
print(f"\nColumns: {df_elec.columns.tolist()}")
print(f"\nData types:\n{df_elec.dtypes}")
print(f"\nFirst 3 rows:\n{df_elec.head(3)}")
print(f"\nMissing values:\n{df_elec.isnull().sum()}")
```
